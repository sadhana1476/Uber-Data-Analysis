```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Month':['Jan','Feb','Mar','Apr','May','Jun'],
    'Trips':[120,150,180,170,210,250]
}

df = pd.DataFrame(data)

df
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(6,4))

plt.plot(df['Month'], df['Trips'])

plt.title("Uber Trips Analysis")

plt.xlabel("Month")

plt.ylabel("Trips")

plt.show()
```

```python
highest = df.loc[df['Trips'].idxmax()]

print("Highest Trips Month:")
print(highest)
```
